## 2.  Проверка ошибок в данных и их предобработка


### 2.1. Названия, или метки, столбцов датафрейма

- Выводим на экран названия всех столбцов датафрейма.
- Приведем все столбцы к стилю snake case. Названия должны быть в нижнем регистре, а вместо пробелов — подчёркивания.

```python
# Вывод названий столбцов
game.columns
```

```python
# перевод названий столбцов к стилю snake case и к нижнему регистру
game.columns = game.columns.str.lower().str.replace(' ', '_')
```

### 2.2. Типы данных



Поле `year_of_release` содержит год выпуска игры и пропуски в данном поле могут свидетельствовать об отсутствии точной информации о дате выхода игры.
Поскольку доля пропусков в столбце `year_of_release` составляет около 1.6% существенного влияния на анализ это не окажет, пропуски в данном поле заменим на 0, чтобы привести к типу данных integer

```python
# заменим пропуски в поле year_of_release на 0
game['year_of_release'] = game['year_of_release'].fillna(0)
```

```python
# приведет поле year_of_release к типу int с понижением разрядности
game['year_of_release'] = pd.to_numeric(game['year_of_release'], downcast='integer')
```

В числовых столбцах могут встретиться строковые значения, например `unknown` или другие. Приведем такие столбцы к числовому типу данных, заменив строковые значения на пропуски.

```python
game['eu_sales'] = pd.to_numeric(game['eu_sales'], errors='coerce')
```

```python
game['jp_sales'] = pd.to_numeric(game['jp_sales'], errors='coerce')
```

```python
# выводим уникальные значения поля `user_score`
game['user_score'].unique()
```

В столбце `user_score` с пользовательскими оценками помимо числовых значений содержатся пропуски - NaN и значение 'tbd', что вероятно расшифровывается как 'to be determined', т.е. оценки не определены. Заменим эти значения на NaN и преведем к типу данных `float`.

```python
game['user_score'] = pd.to_numeric(game['user_score'], errors='coerce')
```

```python
# выведем информацию о датафрейме после преобразования столбцов
game.info()
```

### 2.3. Наличие пропусков в данных

- Считаем количество пропусков в каждом столбце в абсолютных и относительных значениях.


```python
# количество пропусков в абсолютном значении
game.isna().sum()
```

```python
# количество пропусков в относительном значении
game.isna().sum() / game.shape[0]
```

После преобразования типа данных, пропуски содержаться в 7 столбцах.
- `rating` - 6871 пропуск, около 40%. Пропуски вероятно вызваны отсутствием данных, т.к. ESRB - рейтинг для игр, выпущенных на территории США и Канады, для других регионов данный рейтинг может отсутствовать и иметь другую маркировку. Всвязи с этим целесообразно заменить пропущенные значения на 'unknown'.
- `name`, `genre` - по 2 пропуска. Возможные причины пропусков - ошибка ввода или повреждение данных. Их доля составляет 0.01%, что не окажет существенного влияния на анализ данных и строки с пропусками можно удалить.
- `eu_sales` - 6 пропусков и `jp_sales` - 4 пропуска, что составляет 0.03% и 0.02% соответственно. Причина возникновения может быть связана с отсутствием данных о продажах в регионе. Количество пропусков не окажет существенного влияния на анализ данных и строки с пропусками можно удалить.
- `critic_score` и `user_score` содержат 8714 и 9268 пропусков в строках, что составляет 51% и 54% соответственно. Это значительное количество пропусков. Скорее всего это пропуски MNAR — мы не знаем были проведены оценки этих игр или нет. Заменим пропуски уникальным значением(-1)

```python
game['critic_score'] = game['critic_score'].fillna(-1)
game['user_score'] = game['user_score'].fillna(-1)
```

```python
# заменим пропущенные значения на 'unknown'
game['rating'] = game['rating'].fillna('unknown')
```

```python
# Находим все строки, где name, genre, eu_sales, jp_sales имеют пропуска
clean_games = game[game['name'].isnull() | game['genre'].isnull() | game['eu_sales'].isnull() | game['jp_sales'].isnull()]
display(clean_games)
```

```python
# удаляем строки с пропусками в столбцах name, genre и создаем копию датафрейма gamedrop, с которой продолжим работу
gamedrop = game.dropna(subset=['name', 'genre', 'eu_sales', 'jp_sales']).copy()
```

```python
# выводим информацию о датафрейме после обработки пропусков
gamedrop.info()
```

### 2.4. Явные и неявные дубликаты в данных



- Изучим столбцы `platform`, `year_of_release`, `genre`, `rating`.

## 3. Фильтрация данных 

Коллеги хотят изучить историю продаж игр в начале XXI века, и их интересует период с 2000 по 2013 год включительно. Отберем данные по этому показателю и сохраним новый срез данных в отдельном датафрейме `df_actual`.

```python
# фильтруем данные и создаем новый датафрейм df_actual
df_actual = dfgame.copy()
df_actual = df_actual.loc[(df_actual['year_of_release'] >= 2000) & (df_actual['year_of_release'] <= 2013)]
```

```python
# выводим информацию по отфильтрованным данным
df_actual.info()
```

```python
# выводим первые строки отфильтрованной таблицы df_actual
df_actual.head()
```

- Новый, отфильтрованный по годам датафрейм, содержит 12771 строку и 11 столбцов. 6 столбцов типа `float64`, 1 типа `int16` и 4 типа `object`

## 4. Категоризация данных 
    
Проведем категоризацию данных:
- Разделим все игры по оценкам пользователей и выделим такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
# создаем столбец users_category и отражаем в нем разделение оценок пользователей на категории 
df_actual['users_category'] = pd.cut(
    df_actual['user_score'],
    bins=[0, 3, 8, 10],
    labels=['низкая оценка', 'средняя оценка', 'высокая оценка'],
    right=False 
)
```

- Разделим все игры по оценкам критиков и выделим такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
# создаем столбец critics_category и отражаем в нем разделение оценок пользователей на категории
df_actual['critics_category'] = pd.cut(
    df_actual['critic_score'],
    bins=[0, 30, 80, 100],
    labels=['низкая оценка', 'средняя оценка', 'высокая оценка'],
    right=False 
)
```

```python
# сбросим индексы и выведем первые строки измененной таблицы df_actual с новыми столбцами users_category и critics_category
df_actual = df_actual.reset_index(drop=True)
df_actual.head()
```

- После категоризации данных проверим результат: сгруппируем данные по выделенным категориям и посчитаем количество игр в каждой категории.

```python
display("оценки пользователей", df_actual['users_category'].value_counts())
```

```python
display("оценки критиков", df_actual['critics_category'].value_counts())
```
